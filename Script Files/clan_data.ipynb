## **Players Data**

```python
# Define the folder path and ensure it exists
players_folder_path = r'D:\\Python Data\\Projects\\Fukre Work\\Clash of Clan Data\\Players Data'
os.makedirs(players_folder_path, exist_ok=True)
game_folder_path = r'D:\\Python Data\\Projects\\Fukre Work\\Clash of Clan Data'
os.makedirs(game_folder_path, exist_ok=True)

# Fetch clan data
url = f"{BASE_URL}/clans/%23{clan_tag}"
headers = {"Authorization": f"Bearer {API_TOKEN}"}

response = requests.get(url, headers=headers)
if response.status_code == 200:
    clan_data = response.json()
else:
    print(f"Error fetching clan data: {response.status_code} - {response.text}")
    clan_data = None

# Convert clan data to DataFrame
if clan_data:
    members = clan_data.get('memberList', [])
    data = []
    for member in members:
        # Replace 'admin' with 'Elder' in Role
        role = member.get('role', '')
        if role == 'admin':
            role = 'elder'
        
        # Calculate Donation Ratio as percentage
        donated = member.get('donations', 0)
        received = member.get('donationsReceived', 0)
        if received != 0:
            donation_ratio = round((donated / received) * 100, 2)
        else:
            donation_ratio = 0
        
        # Limit ratio to 100% if Troops Donated value is equal to or more than Troops Received
        if donated >= received:
            donation_ratio = 100
        
        data.append({
            "Clan Rank":member.get('clanRank', ''),
            "Player Tag": member.get('tag', ''),
            "Player Name": member.get('name', ''),
            "Role": role,
            "Town Hall Level": member.get('townHallLevel', ''),
            "EXP Level": member.get('expLevel', ''),
            "Home Base Trophies": member.get('trophies', ''),
            "Builder Base Trophies": member.get('builderBaseTrophies', ''),
            "Troops Donated": donated,
            "Troops Received": received,
            "Donation Ratio (%)": donation_ratio
        })
        
    df = pd.DataFrame(data)
    print(df)
    # Save to Excel
    players_excel_filename = os.path.join(players_folder_path, f"{current_date}_clan_players_data.xlsx")
    df.to_excel(players_excel_filename, index=False)

    # Sort the consolidated DataFrame by 'Name' in a case-insensitive manner
    df1 = df.sort_values(by='Player Name', key=lambda x: x.str.lower())
    players_excel_filename1 = os.path.join(game_folder_path, "Players_Data.xlsx")
    df1.to_excel(players_excel_filename1, index=False)
    print(f"Raid data saved to {current_date}_clan_players_data.xlsx")
else:
    # Print error message if request failed
    print(f"Error fetching data: {response.status_code} - {response.text}")
```

## **Raid Weekend Data**

#### *Excel File*

```python
# Define the folder path and ensure it exists
raid_folder_path = r'D:\\Python Data\\Projects\\Fukre Work\\Clash of Clan Data\\Raid Weekend Data'
os.makedirs(raid_folder_path, exist_ok=True)

# Endpoint URL
url = f"{BASE_URL}/clans/%23{clan_tag}/capitalraidseasons"
headers = {"Authorization": f"Bearer {API_TOKEN}"}

# Send GET request to the API endpoint
response = requests.get(url, headers=headers)

# Check if request was successful (status code 200)
if response.status_code == 200:
    # Parse response JSON data
    data = response.json()
    
    # Extract members data
    members_data = []
    for item in data.get("items", []):
        start_time_utc = datetime.strptime(item.get("startTime"), "%Y%m%dT%H%M%S.%fZ")
        end_time_utc = datetime.strptime(item.get("endTime"), "%Y%m%dT%H%M%S.%fZ")
        
        # Convert start and end times to IST
        ist = pytz.timezone("Asia/Kolkata")
        start_time_ist = start_time_utc.replace(tzinfo=pytz.utc).astimezone(ist)
        end_time_ist = end_time_utc.replace(tzinfo=pytz.utc).astimezone(ist)
        
        for member in item.get("members", []):
            # Calculate week number based on start time
            week_number = start_time_ist.isocalendar()[1]
            
            member_data = {
                "Start Time (IST)": start_time_ist.strftime("%Y-%m-%d %H:%M:%S"),
                "End Time (IST)": end_time_ist.strftime("%Y-%m-%d %H:%M:%S"),
                "Week Number": week_number,
                "Player Tag": member.get("tag"),
                "Player Name": member.get("name"),
                "Attacks": member.get("attacks"),
                "Attack Limit": member.get("attackLimit"),
                "Bonus Attack Limit": member.get("bonusAttackLimit"),
                "Capital Resources Looted": member.get("capitalResourcesLooted")
            }
            members_data.append(member_data)
    
    # Convert members data to DataFrame
    df = pd.DataFrame(members_data)
    print(df)
    # Save to Excel
    excel_filename = os.path.join(raid_folder_path, f"{week_number}_raid_weekend_data.xlsx")
    df.to_excel(excel_filename, index=False)
    print(f"Raid data saved to {week_number}_raid_weekend_data.xlsx")
else:
    # Print error message if request failed
    print(f"Error fetching data: {response.status_code} - {response.text}")
```

## **Current War Data**

#### *Json File + Excel File*

```python
# Define the folder path and ensure it exists
current_war_folder_path = r'D:\\Python Data\\Projects\\Fukre Work\\Clash of Clan Data\\Current War Data'
os.makedirs(current_war_folder_path, exist_ok=True)

# Endpoint URL
url = f"{BASE_URL}/clans/%23{clan_tag}/currentwar"
headers = {"Authorization": f"Bearer {API_TOKEN}"}

# Send GET request to the API endpoint
response = requests.get(url, headers=headers)

# Check if request was successful (status code 200)
if response.status_code == 200:
    # Parse response JSON data
    data = response.json()
    print(json.dumps(data, indent=4, ensure_ascii=False))

    # Check the war state
    state = data.get("state")
    if state == "notInWar":
        print("Clan not in daily war")
    else:
        # Extract the preparation start time
        preparation_start_time = data.get("preparationStartTime")
        
        # Convert to IST
        def convert_to_ist(iso_time):
            utc_time = datetime.strptime(iso_time, "%Y%m%dT%H%M%S.%fZ")
            utc_time = utc_time.replace(tzinfo=pytz.utc)
            ist_time = utc_time.astimezone(pytz.timezone('Asia/Kolkata'))
            return ist_time.strftime("%Y-%m-%d")
        
        preparation_start_time_ist = convert_to_ist(preparation_start_time)
        
        # Define file name
        file_name = os.path.join(current_war_folder_path, f"{preparation_start_time_ist}_current_war_data.json")
        
        # Save the data to a JSON file
        with open(file_name, "w", encoding="utf-8") as json_file:
            json.dump(data, json_file, indent=4, ensure_ascii=False)
            print(f"Current war data saved to '{file_name}'")
        
        # Extract members data
        members_data = []
        for member in data["clan"]["members"]:
            member_info = {
                "War Start Date": preparation_start_time_ist,
                "Tag": member.get("tag"),
                "Name": member.get("name"),
                "Townhall Level": member.get("townhallLevel"),
                "Map Position": member.get("mapPosition"),
                "Attack 1 Defender Tag": '',
                "Attack 1 Stars": '0',
                "Attack 1 Destruction Percentage": '0',
                "Attack 1 Duration": '0',
                "Attack 2 Defender Tag": '',
                "Attack 2 Stars": '0',
                "Attack 2 Destruction Percentage": '0',
                "Attack 2 Duration": '0'
            }
            
            attacks = member.get("attacks", [])
            if attacks:
                for i, attack in enumerate(attacks):
                    if i == 0:
                        member_info["Attack 1 Defender Tag"] = attack.get("defenderTag", '')
                        member_info["Attack 1 Stars"] = attack.get("stars", '')
                        member_info["Attack 1 Destruction Percentage"] = attack.get("destructionPercentage", '')
                        member_info["Attack 1 Duration"] = attack.get("duration", '')
                    elif i == 1:
                        member_info["Attack 2 Defender Tag"] = attack.get("defenderTag", '')
                        member_info["Attack 2 Stars"] = attack.get("stars", '')
                        member_info["Attack 2 Destruction Percentage"] = attack.get("destructionPercentage", '')
                        member_info["Attack 2 Duration"] = attack.get("duration", '')
            
            members_data.append(member_info)
        
        # Convert members data to DataFrame
        df = pd.DataFrame(members_data)
        # Sort the DataFrame by 'Map Position' in ascending order
        df = df.sort_values(by="Map Position")
        print(df)
        
        # Save to Excel
        excel_file_name = os.path.join(current_war_folder_path, f"{preparation_start_time_ist}_current_war_data.xlsx")
        df.to_excel(excel_file_name, index=False)
        print(f"Raid data saved to '{excel_file_name}'")
else:
    # Print error message if request failed
    print(f"Error fetching data: {response.status_code} - {response.text}")
```

## *Data Processing*

### OverAll Raid Data

```python
import pandas as pd
import os
import glob

# Directory containing the raid weekend data files
input_folder = r'D:\\Python Data\\Projects\\Fukre Work\\Clash of Clan Data\\Raid Weekend Data'
output_file = r'D:\\Python Data\\Projects\\Fukre Work\\Clash of Clan Data\\Raid_data.xlsx'

# Get all Excel files in the directory
excel_files = glob.glob(os.path.join(input_folder, '*.xlsx'))

# Initialize an empty DataFrame to hold the consolidated data
consolidated_df = pd.DataFrame()

# Iterate over each Excel file
for file in excel_files:
    # Read the Excel file into a DataFrame
    df = pd.read_excel(file)
    
    # Extract the week number from the file name
    week_number = os.path.basename(file).split('_')[0]
    
    # Create a new column name for attacks based on the week number
    new_column_name = f'Week: {week_number}'
    
    # Select the necessary columns and rename the attacks column
    df = df[['Player Tag', 'Player Name', 'Attacks']].rename(columns={'Attacks': new_column_name})
    
    # Merge the current file's data with the consolidated DataFrame
    if consolidated_df.empty:
        consolidated_df = df
    else:
        consolidated_df = pd.merge(consolidated_df, df, on=['Player Tag', 'Player Name'], how='outer')

# Sort the consolidated DataFrame by 'Player Name' in a case-insensitive manner
consolidated_df = consolidated_df.sort_values(by='Player Name', key=lambda x: x.str.lower())

# Write the consolidated data to a new Excel file
consolidated_df.to_excel(output_file, index=False)

print(f"Consolidated data saved to {output_file}")
```

### OverAll Daily War Data

```python
import pandas as pd
import os
import glob

# Directory containing the current war data files
input_folder = r'D:\\Python Data\\Projects\\Fukre Work\\Clash of Clan Data\\Current War Data'
output_file = r'D:\\Python Data\\Projects\\Fukre Work\\Clash of Clan Data\\War_data.xlsx'

# Get all Excel files in the directory
excel_files = glob.glob(os.path.join(input_folder, '*.xlsx'))

# Initialize an empty list to hold the dataframes
dataframes = []

# Iterate over each Excel file
for file in excel_files:
    # Read the Excel file into a DataFrame
    df = pd.read_excel(file)
    
    # Extract the war start date from the file name
    war_start_date = os.path.basename(file).split('_')[0]
    
    # Select the necessary columns
    df = df[['Tag', 'Name', 'Attack 1 Stars', 'Attack 1 Destruction Percentage', 'Attack 2 Stars', 'Attack 2 Destruction Percentage']]
    
    # Rename the attack columns to include the war start date
    df = df.rename(columns={
        'Attack 1 Stars': f'{war_start_date} Attack 1 Stars',
        'Attack 1 Destruction Percentage': f'{war_start_date} Attack 1 DP',
        'Attack 2 Stars': f'{war_start_date} Attack 2 Stars',
        'Attack 2 Destruction Percentage': f'{war_start_date} Attack 2 DP'
    })
    
    # Append the DataFrame to the list
    dataframes.append(df)

# Merge all dataframes on 'Tag', 'Name', and 'Townhall Level'
consolidated_df = dataframes[0]
for df in dataframes[1:]:
    consolidated_df = pd.merge(consolidated_df, df, on=['Tag', 'Name'], how='outer')

# Sort the consolidated DataFrame by 'Name' in a case-insensitive manner
consolidated_df = consolidated_df.sort_values(by='Name', key=lambda x: x.str.lower())

# Write the consolidated data to a new Excel file
consolidated_df.to_excel(output_file, index=False)

print(f"Consolidated data saved to {output_file}")
```
